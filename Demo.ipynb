# EDA 


```python
numericalClient= client[['NO_OF_TRXNS202005','NO_OF_TRXNS202006','TURNOVER202005','TURNOVER202006']]
```

```python
# correlational matrix
corrmat= numericalClient.corr()
f, ax = plt.subplots(figsize=(8, 7))
sns.heatmap(corrmat,annot = True,fmt='.2f',annot_kws={'size':10}, vmax= 1.0, square= True);
```

#### 1.  Which customer segment is most prevalent?


```python
grouped_by_customerseg= client.groupby('Sub Seg')
clientPer = pd.DataFrame(grouped_by_customerseg.size()/client.shape[0]*100).reset_index().rename(columns= {0: 'Segment%'})
plt.figure(figsize=(8,5))
sns.barplot(x= clientPer['Sub Seg'], y= clientPer['Segment%'])
plt.show()
```

Gennesis is most prevalent than the others.


#### 2. Which insurance group is most prevalent?


```python
grouped_by_packname= client.groupby('Insur PackName')
packPer = pd.DataFrame(grouped_by_packname.size()/client.shape[0]*100).reset_index().rename(columns= {0: 'Product Group%'})
plt.figure(figsize=(60,20))
sns.barplot(x= packPer['Insur PackName'], y= packPer['Product Group%'])
plt.show()
```

Funeal is the most prevalent insurance than the other.


#### 3. What is the status of product distribution?


```python
client['Product Status'].value_counts().plot.pie(explode =[0.05]*4,shadow=True,autopct='%1.1f%%',figsize=(12,8),fontsize = 15, labels = None)
                                                 

labels = client['Product Status'].value_counts().index.tolist()
plt.title('Distribution of Product Status')
plt.legend(bbox_to_anchor = (0.85,1),loc = 'upper left',labels = labels)
```

It's possible to notice that we have a large amount of active products and we can consider that the canceled products are a little more than half of the active products.

#### 4. Which customer segment has the most cancellations?


```python
cancelled_data = client[client['Product Status'] == 'CANCELLED']
cancel_grp = cancelled_data.groupby('Sub Seg')
D1 = pd.DataFrame(cancel_grp.size()).rename(columns= {0:'total cancelled per segment'})

grouped_by_segment = client.groupby('Sub Seg')
total_segments = grouped_by_segment.size()
D2 = pd.DataFrame(total_segments).rename(columns ={0:'total clients per segment'})
D3= pd.concat([D1,D2], axis = 1)

D3['cancel (%)']= round((D3['total cancelled per segment']/D3['total clients per segment'])*100,2)
D3
```

```python
plt.figure(figsize = (10,5))
sns.barplot(x = D3.index, y = D3['cancel (%)'])
plt.show
```

Almost 40 % of Gennesis segment got canceled

```python
cancelled_d = client[client['Product Status'] == 'CANCELLED']
cancel_g = cancelled_d.groupby('Insur PackName')
C1 = pd.DataFrame(cancel_g.size()).rename(columns= {0:'total cancelled per product group'})

grouped_by_packname = client.groupby('Insur PackName')
total_packnames = grouped_by_packname.size()
C2 = pd.DataFrame(total_packnames).rename(columns ={0:'total clients per group'})
C3= pd.concat([C1,C2], axis = 1)

C3['cancel (%)']= round((C3['total cancelled per product group']/C3['total clients per group'])*80,2)
C3
```

```python
plt.figure(figsize = (45,15))
sns.barplot(x = C3.index, y = C3['cancel (%)'])
plt.show
```
